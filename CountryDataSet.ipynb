```python
import math

def plot_all_histograms(df, title_prefix=""):
    num_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols) / n_cols)

    plt.figure(figsize=(5 * n_cols, 4 * n_rows))

    for i, col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f"{title_prefix}{col}")
        plt.xlabel("")
        plt.ylabel("")

    plt.tight_layout()
    plt.show()
```

```python
plot_all_histograms(df
                   )
```

```python
df2 = df.drop("country", axis = 1)
```

```python
df2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df.columns
```

```python
df2 = scaler.fit_transform(df2)
```

```python
df2 = pd.DataFrame(df2, columns=['child_mort', 'exports', 'health', 'imports', 'income',
       'inflation', 'life_expec', 'total_fer', 'gdpp'])
```

```python
plot_all_histograms(df2)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca_df2 = pd.DataFrame(pca.fit_transform(df2))
```

```python
pca_df2
```

```python
pca.explained_variance_
```

```python
plt.step(list(range(1,10)),np.cumsum(pca.explained_variance_ratio_))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.ylabel("Variance Covered")
plt.title("Variance Covered")
plt.show()
```

```python
pca_df2 = pca_df2.drop(columns= [3,4,5,6,7,8])
```

```python
pca_df2
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init = "k-means++")
    kmeans.fit(pca_df2)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
plt.xticks(range(1,11))
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters = 3)
```

```python
model.fit(pca_df2)
```

```python
labels = model.labels_
```

```python
df
```

```python
silhouette_score(pca_df2, labels)
```

```python
df['Class'] = labels
```

```python
df
```

```python
fig, ax = plt.subplots(nrows=  1, ncols=2, figsize = (15,5))

plt.subplot(1,2,1)
sns.boxplot(data = df, x = "Class", y = "child_mort")
plt.title("Child mort vs Class")

plt.subplot(1,2,2)
sns.boxplot(data = df, x = "Class", y = "income")
plt.title("income vs Class")

plt.show()
```

```python
import plotly.express as px
```

```python
pca_df2.insert(0, column="Country", value = df["country"])
```

```python
pca_df2['Class'] = labels
```

```python
pca_df2.loc[pca_df2['Class'] == 0, 'Class'] = "Budget Needed"
pca_df2.loc[pca_df2['Class'] == 1, 'Class'] = "No Budget Needed"
pca_df2.loc[pca_df2['Class'] == 2, 'Class'] = "In Between"
```

```python
pca_df2
```

```python
fig = px.choropleth(
    pca_df2[["Country", "Class"]],
    locationmode= "country names", 
    locations="Country",
    title = "Needed Budget By Country",
    color = pca_df2["Class"],
    color_discrete_map={
                        "Budget Needed" : "Red",
                        "In Between" : "Yellow",
                        "No Budget Needed" : "Green"
        
    })
fig.update_geos(fitbounds = "locations", visible = True)
fig.show()
```
